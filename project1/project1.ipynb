```python
from tkinter import *
from tkcalendar import DateEntry

#Başlık bölümü

master = Tk()

otomation = Canvas(master, height=600, width=750)
otomation.pack()

upper_zone = Frame(master, bg="orange")
upper_zone.place(relx=0.1, rely=0.1, relwidth=0.8, relheight=0.1)

left_zone = Frame(master, bg="#ffe4b5")
left_zone.place(relx=0.1, rely=0.20, relwidth=0.60, relheight=0.5)


right_zone = Frame(master, bg="#ffe4b5")
right_zone.place(relx=0.45, rely=0.20, relwidth=0.45, relheight=0.5)


condition_label = Label(upper_zone, bg="orange", text="İşlem:", font="Verdana 12 bold")
condition_label.pack(padx=10, pady=10, side=LEFT)

condition_option = StringVar(upper_zone)
condition_option.set("\t")


condition_menu = OptionMenu(upper_zone, condition_option, "Hasta Kayıt", "Hasta Sevk", "Hasta ikametgah bilgileri")
condition_menu.pack(padx=10, pady=10, side=LEFT)


date_option= DateEntry(upper_zone, width=12, background="Orange", foreground="black", borderwidth=1, locale="de_DE")
date_option._top_cal.overrideredirect(False)
date_option.pack(padx=10, pady=10, side=RIGHT)

date_label = Label(upper_zone, bg="orange", font="Verdana 12 bold", text="İşlem Arşivleri:" )
date_label.pack(padx=10, pady=10, side=RIGHT)


#Kayıt geçmişi bilgileri


Label(left_zone, text="Kayıt Geçmişi:", bg="#ffe4b5", font="Verdana 10 bold").pack(padx=10, pady=10, anchor=NW)

var = IntVar()

year1 = Radiobutton(left_zone, text="2022 öncesi;", bg="#ffefd5", font="Verdana 10", variable=var,value=1)
year1.pack(padx=15, pady=5, anchor=NW)

year2 = Radiobutton(left_zone, text="2023", bg="#ffefd5", font="Verdana 10", variable=var,value=2)
year2.pack(padx=15, pady=5, anchor=NW)

var1 = IntVar()
time1 = Checkbutton(left_zone, text="Geçen Ay", offvalue=0, onvalue=1, bg="#ffefd5", font="Verdana 10", variable=var1)
time1.pack(padx=25, pady=2, anchor=NW)

var2 = IntVar()
time2 = Checkbutton(left_zone, text="Geçen Hafta", offvalue=0, onvalue=1, bg="#ffefd5", font="Verdana 10", variable=var2)
time2.pack(padx=25, pady=2, anchor=NW)

var3 = IntVar()
time3 = Checkbutton(left_zone, text="Bugün", offvalue=0, onvalue=1, bg="#ffefd5", font="Verdana 10", variable=var3)
time3.pack(padx=25, pady=2, anchor=NW)



#Randevu bilgileri

Label(right_zone, text="Mevcut Randevu Bilgileri", bg="#ffe4b5", font="Verdana 10 bold").pack(padx=10, pady=10, anchor=NW)

var = IntVar()

appoint1 = Radiobutton(right_zone, text="Bu Hafta;", bg="#ffefd5", font="Verdana 10", variable=var,value=1)
appoint1.pack(padx=15, pady=5, anchor=NW)

appoint2 = Radiobutton(right_zone, text="Gelecek Hafta;", bg="#ffefd5", font="Verdana 10", variable=var,value=2)
appoint2.pack(padx=15, pady=5, anchor=NW)


varA = IntVar()
day1 = Checkbutton(right_zone, text="Pazartesi", offvalue=0, onvalue=1, bg="#ffefd5", font="Verdana 10", variable=varA)
day1.pack(padx=25, pady=2, anchor=NW)

varB = IntVar()
day2 = Checkbutton(right_zone, text="Salı", offvalue=0, onvalue=1, bg="#ffefd5", font="Verdana 10", variable=varB)
day2.pack(padx=25, pady=2, anchor=NW)

varC = IntVar()
day3 = Checkbutton(right_zone, text="Çarşamba", offvalue=0, onvalue=1, bg="#ffefd5", font="Verdana 10", variable=varC)
day3.pack(padx=25, pady=2, anchor=NW)

varD = IntVar()
day4 = Checkbutton(right_zone, text="Perşembe", offvalue=0, onvalue=1, bg="#ffefd5", font="Verdana 10", variable=varD)
day4.pack(padx=25, pady=2, anchor=NW)

varE = IntVar()
day5 = Checkbutton(right_zone, text="Cuma", offvalue=0, onvalue=1, bg="#ffefd5", font="Verdana 10", variable=varE)
day5.pack(padx=25, pady=2, anchor=NW)


#pdf butonu

from tkinter import messagebox

def pdf():
    son_mesaj= ""
    master.destroy()
    return

gonder_butonu = Button(right_zone, text="PDF İndir", command=pdf)
gonder_butonu.pack(anchor=E)


master.mainloop()
```
